```python
alice=nltk.corpus.gutenberg.words("carroll-alice.txt")
alice
```

```python
alice_norm=[word_al.lower() for word_al in alice if word_al.isalpha()]
```

```python
alice_tags=nltk.pos_tag(alice_norm,tagset="universal")
alice_tags
```

```python
alice_cfd=nltk.ConditionalFreqDist(alice_tags)
```

```python
alice_cfd
```

```python
alice_cfd["over"],alice_cfd["alice"],alice_cfd["whose"]
```

```python
story=nltk.corpus.gutenberg.words('bryant-stories.txt')
```

```python
tags=nltk.pos_tag(story,tagset="universal")
```

```python
tags[:10]
```

```python
for ((word1,tag1),(word2,tag2),(word3,tag3))  in nltk.trigrams(tags):
    if tag1=='NOUN' and word2=='or' and tag3=='NOUN':
        print word1 + " " + word2+ " "+word3
    
```
